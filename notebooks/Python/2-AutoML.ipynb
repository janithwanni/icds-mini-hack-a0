```python
import pandas as pd
import numpy as np
from autogluon.tabular import TabularPredictor
```

```python
directory = '../../data'  # directory where you have downloaded the data CSV files from the competition
label = 'app_status'  # name of target variable to predict in this competition
eval_metric = 'roc_auc'  # Optional: specify that competition evaluation metric is AUC
save_path = directory + '/AutoGluonModels/'  # where to store trained models
```

```python
train_data = pd.read_csv(f"{directory}/train_data.csv")
```

```python
predictor = TabularPredictor(label=label, eval_metric=eval_metric, path=save_path, verbosity=3).fit(
    train_data, presets='best_quality', time_limit=3600
)

results = predictor.fit_summary()
```

```python
# test_identity = pd.read_csv(directory+'test_data.csv')
# test_transaction = pd.read_csv(directory+'test_transaction.csv')
# test_data = pd.merge(test_transaction, test_identity, on='TransactionID', how='left')  # same join applied to training files

test_data = pd.read_csv(f"{directory}/test_data.csv")

y_predproba = predictor.predict_proba(test_data, as_multiclass=False).astype(int)
y_predproba.head(5)
```

```python
submission = pd.read_csv(directory+'/sampleSubmissionFile.csv')
submission[label] = y_predproba
# submission.head()
submission.to_csv(directory+'/automl_submission.csv', index=False)
```
